```python
import datetime as dt
import dt.time as tm
def mk_hr_min(stime):
    (hr,min) = [ int(x) for x in stime.split(':') ]
    
def tdiff(s1, s2):
    h1,m1 = mk_hr_min(s1)
    h2,m2 = mk_hr_min(s2)
    t1 = tm(h1,m1,0)
    t2 = tm(h2,m2,0)
    td = t2 - t1
    return 60*td.hours + td.minutes
    
    
```
