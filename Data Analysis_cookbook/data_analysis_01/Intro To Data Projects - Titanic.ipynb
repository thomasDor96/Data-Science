```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic_df = pd.read_csv('titanic_data.csv')
titanic_df.head()
```

```python
titanic_df.info()
```

```python
sns.factorplot('Sex', data=titanic_df, kind='count')
```

```python
sns.factorplot('Sex', data=titanic_df, hue='Pclass', kind='count')
```

```python
# Interesting find --> Males in 3rd Class >> Females in 3rd Class 
sns.factorplot('Pclass', data=titanic_df, hue='Sex', kind='count')
```

```python
def male_female_child(passenger):
    age, sex = passenger
    if age < 16:
        return 'child'
    else:
        return sex
```

```python
titanic_df['person'] = titanic_df[['Age', 'Sex']].apply(male_female_child, axis=1)
```

```python
titanic_df[0:10]
```

```python
# Not a lot of children travelling First Class
sns.factorplot('Pclass', data=titanic_df, hue='person', kind='count')
```

```python
titanic_df['Age'].hist(bins=80)
```

```python
# Mean Age of passengers
titanic_df['Age'].mean()
```

```python
titanic_df['person'].value_counts()
```

```python
fig = sns.FacetGrid(titanic_df, hue='Sex', aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df, hue='person', aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
fig = sns.FacetGrid(titanic_df, hue='Pclass', aspect=4)
fig.map(sns.kdeplot, 'Age', shade=True)
oldest = titanic_df['Age'].max()
fig.set(xlim=(0,oldest))
fig.add_legend()
```

```python
deck = titanic_df['Cabin'].dropna()
deck.head()
```

```python
levels = []
for level in deck:
    levels.append(level[0])
cabin_df = DataFrame(levels)
cabin_df.columns = ['Cabin']
#sns.factorplot('Cabin', data=cabin_df, palette='winter_d', kind='count')
sns.factorplot('Cabin',data=cabin_df,kind='count',palette='inferno')
```

```python
cabin_df = cabin_df[cabin_df.Cabin != 'T']
sns.factorplot('Cabin',data=cabin_df,kind='count',palette='summer')
```

```python
# Almost all people at Queenstown were 3rd Class
sns.factorplot('Embarked', data=titanic_df, hue='Pclass', kind='count', palette='winter_d', order=['C', 'Q', 'S'])
```

```python
titanic_df['Alone'] = titanic_df.SibSp + titanic_df.Parch
titanic_df.head()
```

```python
titanic_df['Alone'].loc[titanic_df['Alone'] > 0] = 'With Family'
titanic_df['Alone'].loc[titanic_df['Alone'] == 0] = 'Alone'
```

```python
titanic_df.head()
```

```python
sns.factorplot('Alone', data=titanic_df, palette='Blues', kind='count')
```

```python
titanic_df['Survivor'] = titanic_df.Survived.map({0: 'No', 1: 'Yes'})
```

```python
sns.factorplot('Survivor', data=titanic_df, palette='Set1', kind='count')
```

```python
sns.factorplot('Pclass', 'Survived', hue='person', data=titanic_df)
```

```python
sns.lmplot('Age', 'Survived', data=titanic_df)
```

```python
sns.lmplot('Age', 'Survived', hue='Pclass', palette='winter', data=titanic_df)
```

```python
generations = [10, 20, 30, 40, 50, 60, 70, 80]
sns.lmplot('Age', 'Survived', hue='Pclass', palette='winter', data=titanic_df, x_bins=generations)
```

```python
sns.lmplot('Age', 'Survived', hue='Sex', data=titanic_df, palette='winter', x_bins=generations)
```
