## Darknet Dataset Parsing To Find Sum of All Fentanyl Compounds 

```python
import pandas as pd
```

```python
# Loading the Excel file
xlsfile = pd.ExcelFile('Trial of Word List Version 3 -Based on 21 Oct 2017 Data for Dream Market.xlsx')
```

```python
# Loading the sheet into a dataframe
dframe = xlsfile.parse('csv-zusammenfuehren.de_lkpmk5nk')
dframe.head()
```

```python
# Isolating the rows containing 'Fentanyl' into another dataframe
dfFent = dframe[dframe['Product:'].str.contains("fentanyl")==True]
dfFent.head()
```

```python
# Creating a dictionary of units and their multiplication factor
d = {'gm':1, 'gr':1, 'grams':1, 'kg':1000, '%':.01, 'g':1, 'kilograms':1000, '%pure':0.01, 'mcg':0.0000001}
```

```python
# Isolating the numeric figures and the weights/purity.
dfWts =  dfFent['Product:'].str.extractall('(?P<Qty>\d+\.\d+|\d+)\s*(?P<Unit>[a-z%]+)')
print(dfWts.head())
```

```python
# Multiplication and summation over all rows of data...
sumWts = dfWts['Qty'].astype(float).mul(dfWts['Unit'].map(d)).unstack().prod(axis=1).sum()
print (sumWts, ' grams of Fentanyl or similar compounds.')
```

```python
# Dataframe to Excel Export
writer = pd.ExcelWriter('PythonExport.xlsx')
dfWts.to_excel(writer,'Sheet1')
writer.save()
```
