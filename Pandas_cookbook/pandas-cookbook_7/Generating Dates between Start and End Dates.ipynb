```python
df = pd.read_clipboard()
```

```python
df.dtypes
```

```python
# Convert dates to datetime dytpes
df.EndDate = pd.to_datetime(df.EndDate)
df.StartDate = pd.to_datetime(df.StartDate)
```

```python
# Set Index to StartDate
df = df.set_index('StartDate', drop = False)
```

```python
# Create Empty DataFrame
new_df = pd.DataFrame()
```

```python
# Iteration through each ID and Generate Dates between Start and End Dates
for i, data in df.iterrows():
    data = data.to_frame().transpose()
    data = data.reindex(pd.date_range(start=data.index[0], end=data.EndDate[0])).fillna(method='ffill').reset_index().rename(columns={'index': 'Date'})
    new_df = pd.concat([new_df, data])
```
